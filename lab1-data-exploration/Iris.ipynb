### 1. Load the data

```python
# import package
import pandas as pd
```

### 2. Summary statistics

```python
# Summary Statistics
df.describe(include='all')
```

```python
# how to check the mean or the stdev of a column?
print(df['SepalLengthCm'].mean())
print("The mean of SepalLengthCm is", df['SepalLengthCm'].mean())
print("The standard deviation of SepalLengthCm is", df['SepalLengthCm'].std())
print("The median of SepalLengthCm is", df['SepalLengthCm'].median())
```

```python
# (advanced) check the mean of four numeric columns
for col in df.columns[1:-1]:
    print("The mean of {} is {:.2f}".format(col, df[col].mean()))
```

##### Draw a histogram using matplotlib

```python
# import package
import matplotlib.pyplot as plt

# histogram
plt.hist(df['PetalWidthCm'], bins=10) # number of bins: 10

# details
plt.xlabel('PetalWidth')
plt.ylabel('Frequency')
plt.title('Histogram of PetalWidthCm')
```

##### Draw a histogram using seaborn

```python
# import package
import seaborn as sns

# histogram
sns.histplot(df['PetalWidthCm'], bins=10).set(title='Histogram of PetalWidthCm')
```

```python
# for the box plot above, we will use four numeric columns only
df_num = df.iloc[:, 1:-1] # slicing syntax
df_num
```

```python
# boxpot
sns.boxplot(df_num).set(title='Boxplot of Iris')
```

```python
# heatmap
corr = df_num.corr()
sns.heatmap(corr, annot=True).set(title="Correlation of Iris Dataset")
```

```python
# scatter plot
sns.scatterplot(x=df['SepalLengthCm'], 
                y=df['PetalLengthCm']).set(title="Scatter Plot")
```

```python
# scatter plot (hue added)
sns.scatterplot(x=df['SepalLengthCm'], 
                y=df['PetalLengthCm'],
                hue=df['Species']).set(title="Scatter Plot")
```

```python
# scattermatrix
sns.pairplot(df.iloc[:, 1:], hue='Species')
```
