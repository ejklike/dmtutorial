#### Import packages

```python
# jupyter-notebook-specific command:
# shows the plots without `plt.show()` in Jupyter notebook
%matplotlib inline

# import packages
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# increase default figure and font sizes for easier viewing
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['font.size'] = 14
```

In this tutorial, we will explore [Iris dataset](http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data) from the UCI machine learning repository. This data contain information on 150 Iris flowers, 50 each from one of three Iris species: Setosa, Versicolour, and Virginica. Each flower is characterized by five attributes:

- sepal length in centimeters

- sepal width in centimeters

- petal length in centimeters

- petal width in centimeters

- class (Setosa, Versicolour, Virginica) 

**<font color="red">Code:</font>** The following code uses Pandas to read the CSV file and store them in a DataFrame object named `data`. Next, it will display the first five rows of the data frame using `.head()`.


```python
data = pd.read_csv('Iris.csv')
data.head()
```

The following code changes the column names for brevity.

```python
data.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'class']

data.head()
```

## 3.1. Summary Statistics

Summary statistics are quantities, such as the mean and standard deviation, that capture various characteristics of a potentially large set of values with a single number or a small set of numbers. 

In this tutorial, you will learn how to compute various summary statistics from the DataFrame.

To execute the sample program shown here, make sure you have installed the Pandas library (see Module 2).

**1.** For each quantitative attribute, calculate its average, standard deviation, minimum, and maximum values.

**<font color="red">Code:</font>**

```python
for col in ['sepal length', 'sepal width', 'petal length', 'petal width']:
    print('%s:' % (col))
    print('\t Mean = %.2f' % data[col].mean())
    print('\t Standard deviation = %.2f' % data[col].std())
    print('\t Minimum = %.2f' % data[col].min())
    print('\t Maximum = %.2f' % data[col].max())
```

**2.** For the qualitative attribute (class), count the frequency for each of its distinct values.

**<font color="red">Code:</font>**

```python
data['class'].value_counts()
```

**3.** It is also possible to display the summary for all the attributes simultaneously in a table using the describe() function. If an attribute is quantitative, it will display its mean, standard deviation and various quantiles (including minimum, median, and maximum) values. If an attribute is qualitative, it will display its number of unique values and the top (most frequent) values. 

**<font color="red">Code:</font>**

```python
data.describe(include='all')
```

Note that count refers to the number of non-missing values for each attribute.

**4.** For multivariate statistics, you can compute the covariance and correlation between pairs of attributes.

**<font color="red">Code:</font>**

```python
print('Correlation:')
data.corr()
```

## 3.2. Data Visualization

Data visualization is the display of information in a graphic or tabular format. Successful visualization requires that the data (information) be converted into a visual format so that the characteristics of the data and the relationships among data items or attributes can be analyzed or reported.

In this tutorial, you will learn how to display the Iris data. To execute the sample program shown here, make sure you have installed the `matplotlib` library package.

Befor visualization, you need to import the `matplotlib` package.

**1.** First, we will display the histogram for the sepal length attribute by discretizing it into 8 separate bins and counting the frequency for each bin.

**<font color="red">Code:</font>**

```python
plt.hist(data['sepal width'], bins=8)
# data['sepal width'].hist(bins=8) # do the same thing as above
plt.xlabel('sepal width')
plt.ylabel('count')
plt.title('Histogram of sepal width')
plt.show()
```

**2.** A boxplot can also be used to show the distribution of values for each attribute.

**<font color="red">Code:</font>**

```python
plt.boxplot(data.iloc[:, :-1])
# data.boxplot() # do the same thing as above
plt.xlabel('column')
plt.ylabel('value')
plt.title('Box plot of Iris')
plt.show()
```

**3.** For each pair of attributes, we can use a scatter plot to visualize their joint distribution.

**<font color="red">Code:</font>**

```python
x = 'petal width'
y = 'petal length'

plt.scatter(data[x], data[y])

plt.xlabel(x)
plt.ylabel(y)
plt.title('Scatter plot')
plt.show()
```

```python
# Scatter matrix
fig, axes = plt.subplots(4, 4, figsize=(12, 12)) # 4 x 4 plots in a figure

for i in range(4):
    col_i = data.columns[i]
    for j in range(4):
        col_j = data.columns[j]
        if j == i:
            axes[i][j].hist(data[col_i], bins=10)
            axes[i][j].set_xlabel(col_i)
            axes[i][j].set_ylabel('count')
        else:
            axes[i][j].scatter(data[col_i], data[col_j])
            axes[i][j].set_xlabel(col_i)
            axes[i][j].set_ylabel(col_j)
plt.tight_layout()
```

You can obtain the same result with only one line below:

```python
pd.plotting.scatter_matrix(data[data.columns[:4]], figsize=(12, 12))
plt.show()
```

You can save a plot to a file.

```python
data['sepal width'].plot(kind='hist', bins=20, title='Histogram of Sepal Width')
plt.xlabel('Sepal Width')
plt.ylabel('Frequency')
plt.savefig('sepal_width_histogram.png')
```

## Exercise 1: Drinks
data source: https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv

```python
drinks = pd.read_csv('drinks.csv') # , na_filter=False

drink_cols = ['country', 'beer', 'spirit', 'wine', 'liters', 'continent']
drinks.columns = drink_cols

drinks.head()
```

### Basic check

- Column type (Numerical or Categorical)
- Summary statistics
- Missing values (`nan`)
- ...

```python
# summary statistics of all columns
drinks.describe(include='all')
```

```python
drinks['continent'].value_counts() # AFrica, EU, ASia, OCeania, SouthAmerica
```

```python
drinks.isna().sum() # 23 missing values in `continent` column
```

```python
drinks[drinks.continent.isna()] # check the rows with missing values
```

### Histogram
Purpose: Showing the distribution of a numerical variable

```python
# histogram
bins = 10

drinks.beer.plot(kind='hist', bins=bins)
plt.xlabel('Beer Servings')
plt.ylabel('Frequency')
plt.title('Histogram of Beer Servings (nbin=%d)' % bins)
```

```python
# compare with density plot (smooth version of a histogram)
drinks.beer.plot(kind='density', xlim=(0, 400))
```

```python
# histogram of beer servings grouped by continent
drinks.hist(column='beer', by='continent')
# plt.tight_layout()
```

```python
# share the x axes
drinks.hist(column='beer', by='continent', sharex=True)
```

```python
# share the x and y axes
drinks.hist(column='beer', by='continent', sharex=True, sharey=True)
```

```python
# change the layout
drinks.hist(column='beer', by='continent', sharex=True, layout=(2, 3))
```

### Scatter plot
Purpose: Showing the relationship between two numerical variables

```python
# scatter plot
drinks.plot(kind='scatter', x='beer', y='wine')
```

```python
# add transparency
drinks.plot(kind='scatter', x='beer', y='wine', alpha=0.3)
plt.show()
```

```python
# scatter matrix of three numerical columns
columns = ['beer', 'spirit', 'wine']
pd.plotting.scatter_matrix(drinks[columns], figsize=(10, 10))
plt.show()
```

### Bar plot
Purpose: Showing a numerical comparison across different categories

```python
# count the number of countries in each continent
drinks.continent.value_counts()
```

```python
# compare with bar plot
drinks.continent.value_counts().plot(kind='bar')
```

```python
# calculate the mean alcohol amounts for each continent
drinks.groupby('continent').mean()
```

```python
# side-by-side bar plots
drinks.groupby('continent').mean().plot(kind='bar')
```

```python
# drop the liters column
drinks.groupby('continent').mean().drop('liters', axis=1).plot(kind='bar')
```

```python
# stacked bar plots
drinks.groupby('continent').mean().drop('liters', axis=1).plot(kind='bar', stacked=True)
```

### Box plot
Purpose: Showing quartiles and outliers for one or more numerical variables

```python
# box plot of beer servings
drinks.beer.plot(kind='box')
```

```python
# include multiple variables
drinks.drop('liters', axis=1).plot(kind='box')
```

```python
# box plot of beer servings grouped by continent
drinks.boxplot(column='beer', by='continent')
```

```python
# box plot of all numeric columns grouped by continent
drinks.boxplot(by='continent')
```
