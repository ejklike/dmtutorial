```python
import pandas as pd
```

# Concatenate

```python
# load two data tables
df_w1 = pd.read_csv('weather_1.csv')
df_w2 = pd.read_csv('weather_2.csv')
```

```python
df_w1
```

```python
df_w2
```

```python
# concat two tables
df_w12 = pd.concat([df_w1, df_w2], axis=0)
df_w12
```

```python
# if required, remove duplicates
df_w12 = df_w12.drop_duplicates()
```

```python
df_w12
```

```python
# reset index for a better look
df_w12 = df_w12.reset_index() # reset index
df_w12 = df_w12.drop(['index'], axis=1) # drop old index
df_w12
```

# Join

```python
# load another data tables
df_w3 = pd.read_csv('weather_3.csv')
df_w3
```

```python
# join two tables (left join method, key: Date)
df_w123 = pd.merge(df_w12,df_w3, how='left', left_on='Date', right_on='Date')
df_w123
```
