# Load data

```python
import pandas as pd

df = pd.read_csv('coffee.csv')
df
```

# Groupby Operation
How to calculate the average price and calories group by Menu Type?

```python
# we need Type, Price, and Colories columns
df_for_groupby = df.drop(['MenuNumber'], axis=1)
df_for_groupby
```

```python
# mean by menu type
df_for_groupby.groupby('Type').mean()
```

```python
# median by menu type
df_for_groupby.groupby('Type').median()
```

```python
# count by menu type
df_for_groupby.groupby('Type').count()
```
